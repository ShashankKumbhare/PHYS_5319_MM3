## <span style="color:red"> Problem 1 </span>
Test the 5<sup>th</sup> moment of the random numbers sequence in your programming language  
for 10,000 numbers. Does it correlate or not?

### <span style="color:skyblue"> Solution Problem 1</span>
k<sup>th</sup> moment of random sequence of number is given by,
![Euler Method](images/1_sol_kth_moment.png)

and k<sup>th</sup> moment should be approcximately equals to 1/(k+1).

```python
################################################################################
######################## test_moment_k_th_rand_seq #############################
def test_moment_k_th_rand_seq(k, n):
    
    """
    General info:
        This function returns k-th moment of a random sequence for a specific
        sequence length.
    Arguments:
        k : moment
        n : sequence length
    """
    
    import numpy as np
    
    # Generating a random number sequence of length n >>
    seq = np.random.random(k)
    
    # Calculatin value of k-th moment >>
    value_moment = 1 / n * sum( i**5 for i in seq )
    print("Sequence length (n): " + str(n))
    print("Order of moment (k): " + str(k) + "-th")
    print("Value of " + str(k) + "-th moment: " + str(value_moment))
    print("1/(k+1)             : " + str(1/(k+1)))
    
    # Testing k-th moment (10% tolerence) >>
    if value_moment <= 1 / (k+1) * 1.1:
        print("Test PASSED.")
    else:
        print("Test FAILED.")
    
    return None
######################## test_moment_k_th_rand_seq #############################
################################################################################
```

## <span style="color:green"> Final Solution Problem 1</span>

```python
test_moment_k_th_rand_seq( k = 5,
                           n = 10000 )
```

### Now, since the value of 5-th moment (5.3752954859e-05) is much lesser than 1/(k+1) (0.166), we can say that it almost has no correlation at all.

## <span style="color:red"> Problem 2 </span>
Take 500 steps 2-dimensional random walk with a unit step length in a single trial,  
making ∆x<sub>i</sub> & ∆y<sub>i</sub> in a range of (-1,+1).  
(You need to shift/rescal the random numbers in a range [-1,+1])  

a. Plot the x-y path for this walk.  

b. Plot sqrt(R) vs. sqrt(N) (N is the N-th step), where R = sqrt(x<sup>2</sup> + y<sup>2</sup>).  

c. Then average over 100 trial, plot sqrt(R) vs. sqrt(N) again. Discuss.  

### <span style="color:skyblue"> Solution Problem 2</span>

```python
################################################################################
######################## rand_walk #############################################
def rand_walk_2d( n            = 500,
                  point_start  = (0, 0),
                  length_step  = 1,
                  plot_enabled = True
                ):
    
    """
    General info:
        This function plots random walk with a step length of given length in a
        single trial.
    Arguments:
        n            : number of steps
        length_step  : step length
        plot_enabled : plots random walk path if True
    """
    
    import numpy as np
    gen_rand = np.random.random
    
    # Defining x & y positions as lists >>
    positions_x = np.array( [np.nan] * (n+1) )
    positions_y = np.array( [np.nan] * (n+1) )
    
    # Initial position >>
    positions_x[0] = point_start[0]
    positions_y[0] = point_start[1]
    
    # Taking x & y steps of unit length >>
    steps_x = np.array( [ (length_step if np.random.uniform(-1, 1) >= 0 else -1 * length_step) for _ in range(n) ] )
    steps_y = np.array( [ (length_step if np.random.uniform(-1, 1) >= 0 else -1 * length_step) for _ in range(n) ] )
    
    # Calculating x & y positions >>
    for i in range(n):
        positions_x[i+1] = positions_x[i] + steps_x[i]
        positions_y[i+1] = positions_y[i] + steps_y[i]
    
    # Plotting x & y positions >>
    if plot_enabled:
        fig  = plt.figure(figsize = (10, 10))
        axes = plt.gca()
        axes.plot( positions_x, positions_y )
        # Setting plot elements >>
        axes.set_title("Random 2D walk of step length " + str(length_step))
        axes.set_xlabel("X")
        axes.set_ylabel("Y")

    # Calculating N & R:
    N      = np.array( list(range(n+1)) )
    R_sqrt = np.sqrt(positions_x**2 + positions_y**2)
        
    # Plotting sqrt(R) vs N >>
    if plot_enabled:
        fig  = plt.figure(figsize = (10, 10))
        axes = plt.gca()
        axes.plot( N, R_sqrt)
        # Setting plot elements >>
        axes.set_title("sqrt(R) vs N")
        axes.set_xlabel("N")
        axes.set_ylabel("sqrt(R)")
    
    return positions_x, positions_y, N, R_sqrt
######################## rand_walk #############################################
################################################################################
```

### <span style="color:green"> Final Solution Problem 2</span>

```python
import random
print(random.uniform(-1,1))
```

```python
x, y, N, R = rand_walk_2d( n           = 500,
                           point_start = (0, 0),
                           length_step = 1 )
```

### <span style="color:skyblue"> Average over 100 trial</span>

```python
trials      = 1000
n           = 500
point_start = (0, 0)

R_sqrt = [0] * trials
for i in range(trials):
    x, y, N, R_sqrt[i] = rand_walk_2d( n            = n,
                                    point_start  = point_start,
                                    plot_enabled = False )
    
R_average = sum(R_sqrt) / trials
N         = np.array( list( range(n+1) ) )

# Plotting R_average vs N for many trials >>
fig  = plt.figure(figsize = (10, 10))
axes = plt.gca()
axes.plot( N, R_average )

# Setting plot elements >>
axes.set_title("R_average vs N")
axes.set_xlabel("N")
axes.set_ylabel("R_average")
```

## <span style="color:red"> Problem 3 </span>
Use Monte Carlo method to calculate,  
![prob_3_I](images/3_prob_integration.png)  
up to 5 significant figures.  
The analytic value is 155/6.

### <span style="color:skyblue"> Solution Problem 3</span>

```python
################################################################################
######################## integration_monte_carlo ###############################
def integration_monte_carlo( func,
                             a   = 0,
                             b   = 1,
                             d   = 1,
                             n   = 10000,
                             I_a = None
                           ):
    
    """
    General info:
        This function retunrns integration performed using monte-carlo method 
        for the given mathematical function.
    Arguments:
        func : mathematical function
        a    : vector for lower limits,
        b    : vector for higher limits,
        d    : no. of dimentions
        n    : no. of points to use for integration calculation
        I_a  : Analytical integration (for comparision)
    """
    
    import numpy as np
    
    a = np.array(a)
    b = np.array(b)
    
    # Loop for sum of function values >>
    sum_loop = 0
    for i in range(n):
        sum_loop = sum_loop + func( np.random.uniform(a, b, d) )
        # print(sum_loop)
    
    # Calculating product of difference of limits >>
    diff_limits      = np.subtract(b, a)
    prod_diff_limits = np.prod(diff_limits)
    
    # Calculating integral >>
    I_mc = 1 / n * prod_diff_limits * sum_loop
    
    # Printing results >>
    print( "I_mc : " + str(I_mc) )
    print( "I_a  : " + str(I_a) ) if I_a else None
    print( "error: " + str( (I_mc - I_a) / I_a * 100 ) + " %" ) if I_a else None
    
    return I_mc
######################## integration_monte_carlo ###############################
################################################################################
```
